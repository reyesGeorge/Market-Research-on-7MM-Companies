```python
import pandas as pd
import numpy as np
import re
```

```python
# Importing Data and Indexing by Industry
cData = pd.read_csv('7mm_companies.csv', thousands=None)
```

```python
cData.set_index('industry')
```

```python
uniques = cData.industry.value_counts()
```

```python
# Industry Names
# uniques.index.to_list()
```

```python
employees_by_industry = cData[cData['country'] == 'united states'].groupby(['industry'])['current employee estimate'].sum()
```

```python
employees_by_industry.sort_values(ascending = False).head(10)
```

```python
drop_df = cData.dropna()
```

```python
Texas = drop_df[drop_df['locality'].str.contains('san antonio, texas, united states', regex = True, na = False)]
```

```python
#Texas[Texas['current employee estimate'] <= 100].sort_values('current employee estimate', ascending = False).head(20)
```

```python
Texas_range = (Texas.loc[(Texas['current employee estimate'] <= 100) & (Texas['current employee estimate'] >= 9)].sort_values('current employee estimate', ascending = False))
```

```python
Texas_range.head(3)
```

```python
Texas_range.to_csv(r'Local_City_Leads.csv', header=['Account Name', 'Account Website', 'Location'], columns=['name', 'domain', 'locality'], index=False)
```
